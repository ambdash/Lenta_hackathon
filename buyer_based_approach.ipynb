```python
def get_weight(row):
   weight = row['item_freq'] + row['item_count'] 
   return weight
```

```python
def get_top_items(input_data):

  # Сортировка позиций по частоте
  data_freq = pd.DataFrame(input_data['item_id'].value_counts()).reset_index()
  data_freq.rename(columns = {'index': 'item_id', 'item_id' : 'item_freq'}, 
                   inplace = True)

  # Сортировка позиций по количеству
  data_qty = input_data.groupby(['item_id']).agg('sum').reset_index()
  data_qty = data_qty[['item_id', 'count']]
  data_qty = data_qty.sort_values(by = 'count', ascending = False)
  data_qty = data_qty.reset_index(drop = True)
  data_qty.rename(columns = {'item_id': 'item_id', 'count': 'item_count'},
                  inplace = True)

  # Объединение результатов сортировки позиций по частоте
  # и сортировки позиций по количеству
  data_merged = data_freq.merge(data_qty, on = 'item_id')
  data_merged['weight'] = data_merged.apply(get_weight, axis = 1)
  data_merged = data_merged.sort_values(by = 'weight', ascending = False)
  top_items = pd.DataFrame(
      data_merged['item_id'].head(20)).reset_index(drop = True)
  
  return top_items
```

```python
# соберем словарь встречаемостей - какие item_id покупались чаще с 
# каждым item_id

import gc

from collections import Counter 


hist_data = df

tmp = (
    hist_data[['item_id', 'pav_order_id']]
    .sort_values(['item_id', 'pav_order_id'])
    .merge(hist_data[['item_id', 'pav_order_id']], how='left', on=['pav_order_id'], suffixes=('', '_left'))
)
tmp = tmp[tmp['item_id'] != tmp['item_id_left']].copy()
tmp1 = tmp.groupby(['item_id'])['item_id_left'].agg(lambda x: Counter(x).most_common(10))

most_freq_dict = {k: v for (k, v) in tmp1.iteritems()}
```

```python
# из списка кандидатов по совстречаемости удаляем повторяющиеся item_id, 
# сохраняя порядок
def get_unique_recs(recs: list, top_n: int) -> list:
    rec_dict = {}
    counter = 0
    for k, v in recs:
        if k not in rec_dict:
            rec_dict[k] = v
            counter += 1
        if counter == top_n:
            break
    return list(rec_dict.keys())
```

```python
def rec_by_item(item_id: int, most_freq_dict: dict) -> list:
    
    return most_freq_dict.get(item_id, None)
```

```python
# для каждого item_id соберем top_n самых часто встречающихся item_id,
# отсортируем по частоте и выберем уникальные
def rec_by_basket(basket: list, most_freq_dict: dict,
                  top_n: int = 20) -> list:
    
    res = []
    for item in basket:
        recs = rec_by_item(item, most_freq_dict)
        if recs is not None:
            res += recs
    
    res = sorted(res, key=lambda x: x[1], reverse=True)
    
    return get_unique_recs(res, top_n)
```
